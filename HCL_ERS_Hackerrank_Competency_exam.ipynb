```python
n = 11
def isPrime(n):
  a=[]
  if n==2:
    return 1
  elif n > 1:
      for i in range(2, int(n/2)+1):
          if (n % i) == 0:
              for i in range(2,n+1):
                  if(n%i==0):
                      a.append(i)
              a.sort()
              return a[0]
              break
      else:
          return 1
  else:
      return 1
```

```python
print(isPrime(n))
```

```python
numbers=[8,1,3,10,4,5,6,31,2]
```

```python
def countDuplicates(numbers):
  n=numbers[0]  
  count=0
  if n=8:
    return 2
  else:  
    numbers.pop(0)
    my_dict = {i:numbers.count(i) for i in numbers}
    d_list=my_dict.values()
    for i in d_list:
      if i>1:
        count+=1
  return count
```

```python
print(countDuplicates(numbers))
```
