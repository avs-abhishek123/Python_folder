Just to check the logic of our main code, we use these inputs

```python
n=5
m=4
s="abaca"
x=[1,1,3,4]
y=[2,3,4,5]
```

Our main code

```python
def beauty(n,m,s,x,y):
  l=[]
  s_index_list=[]
  s_list=list(map(str, s))
  count = 0
  maximum=0

  string_list=[]
  l=list(map(max, zip(x, y)))
  l[0] = x[0]
  for i in l:
    ele=i-1
    s_index_list.append(ele)

  s_set = set(s_index_list)
  dup = len(s_index_list) != len(s_set)
  if dup==True:
    return -1
  else:
    for i in s_index_list:
      string_list.append(s_list[i])
    
    length=len(string_list)
    for i in range(0,length-1):
        count=1
        for j in range(i+1,length):
          if(string_list[i]==string_list[j]):
              count+=1
              if(maximum<count):
                maximum=count
    return maximum
```

```python
print(beauty(n,m,s,x,y))
```
